```python
import os
import requests as req
from bs4 import BeautifulSoup as bs
import bs4
import pickle as pkl
```

```python

for num in range(2764,2765):
    try:
        url = "https://www.xkcd.com//{}/".format(num)
        site = bs(req.get(url, allow_redirects=True).text, "html.parser")
        cooldiv = site.find(id="middleContainer")
        links = cooldiv.find_all('a', href=True)
        cool = ''
        for aaa in links:
            if 'imgs.xkcd.com' in aaa['href']:
                cool = aaa['href']
        words = cool[29:-4].split('_')
        final = []
        for word in words:
            if not ('(' in word or ')' in word or 'cropped' in word):
                final.append(word)
        biglist.append([final])
    except:
        print(num)
print(biglist)
```
