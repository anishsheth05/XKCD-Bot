```python
import os
import requests as req
from bs4 import BeautifulSoup as bs
import bs4
import pickle as pkl
```

```python
num = 24
url = "https://www.explainxkcd.com/wiki/index.php/{}".format(num)
site = bs(req.get(url, allow_redirects=True).text, "html.parser")
site
```

```python
transcript_header = site.find(id="Transcript").parent
transcript_header
```

```python
def isHeader(tag: bs4.element.Tag) -> bool:
    if "h2" in str(tag):
        return True
    else:
        return False
```
