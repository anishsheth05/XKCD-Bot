```python
import os
import requests
from bs4 import BeautifulSoup
import pickle as pkl


def search(stuff: list, criteria: str):
    passing = []
    for item in stuff:
        if criteria in str(item):
            passing.append(item)
    return passing



theres_more = True
num = 1038

while theres_more:
    if os.path.exists("images/comic_{}.png".format(num)):
        pass
    else:
        url = "https://xkcd.com/{}/".format(num)
        try:
            site = requests.get(url)
            soup = BeautifulSoup(site.text, 'html.parser')
            a_list = soup.find_all(name="a")
            pic_link = search(search(a_list, "<a href"), "https://imgs.xkcd.com/comics")[0].contents[0]
            print(pic_link)
            with open('images/comic_{}.png'.format(num), 'wb') as file:
                file.write(requests.get(pic_link, allow_redirects=True).content)
        except Exception as e:
            print(e)
            theres_more = False
    num += 1



```
